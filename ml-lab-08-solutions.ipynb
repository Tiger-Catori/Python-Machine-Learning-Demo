## Plotting a few distributions

This example uses matplotlib only, but you can use seaborn too.


```python
# example of plotting a histogram of a random sample
import numpy as np
import matplotlib.pyplot as plt
import seaborn

n = 100000 # used to set how many samples for the distributions created below
b = 50 # used to set the number of bins in the histogram plots below
fig, axs = plt.subplots(3, 2, figsize=(16, 16), sharey=False)

# generate a normal sample
sample_normal = np.random.normal(size=n)
axs[0,0].hist(sample_normal, bins=b, density=True)
axs[0,0].set_title('Gaussian/Normal')

# generate a poisson sample
sample_poisson = np.random.poisson(lam=15, size=n)
axs[0,1].hist(sample_poisson, bins=b, density=True)
axs[0,1].set_title('Poisson')

# generate a uniform sample
sample_uniform = np.random.uniform(size=n)
axs[1,0].hist(sample_uniform, bins=b, density=True)
axs[1,0].set_title('Uniform')

# generate a exponential sample
sample_exponential = np.random.exponential(size=n)
axs[1,1].hist(sample_exponential, bins=b, density=True)
axs[1,1].set_title('Exponential')

# generate a logistic sample
sample_logistic = np.random.logistic(size=n)
axs[2,0].hist(sample_logistic, bins=b, density=True)
axs[2,0].set_title('Logistic')

# generate a power sample
sample_power = np.random.power(a=5., size=n)
axs[2,1].hist(sample_power, bins=b, density=True)
axs[2,1].set_title('Power')
```

## Checking for normality

```python
# importing 3 different tests for normality
from scipy.stats import shapiro     # Shapiro Wilk
from scipy.stats import normaltest  # D’Agostino’s K^2
from scipy.stats import anderson    # Anderson-Darling

print("Normality tests (of normal distribution)")
stat, p = shapiro(sample_normal)
print(' - Shapiro:    W = %0.2f, p-value = %0.2f' % (stat, p))
stat, p = normaltest(sample_normal)
print(' - D’Agostino: W = %0.2f, p-value = %0.2f' % (stat, p))

# bonus test from https://machinelearningmastery.com/a-gentle-introduction-to-normality-tests-in-python/
result = anderson(sample_normal)
print(' - Anderson-Darling statistic: %.3f' % result.statistic)
p = 0
for i in range(len(result.critical_values)):
	sl, cv = result.significance_level[i], result.critical_values[i]
	if result.statistic < result.critical_values[i]:
		print('    %.3f: %.3f, data looks normal (fail to reject H0)' % (sl, cv))
	else:
		print('    %.3f: %.3f, data does not look normal (reject H0)' % (sl, cv))

print()
```

## Estimating distributions

### Parametric

Example with normal distribution.


```python
# example of plotting a histogram of a random sample
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# generate a sample with a mean of 50 and standard deviation of 5
sample = np.random.normal(loc=50, scale=5, size=1000)

# calculate parameters
sample_mean = np.mean(sample)
sample_std = np.std(sample)
print('Mean = %.3f, Standard Deviation = %.3f' % (sample_mean, sample_std))

# define/fit the distribution - assuming it's normal (even though we do know it is...)
distribution = norm(sample_mean, sample_std) # norm = function from SciPy

# sample probabilities for a range of outcomes
values = [value for value in range(30, 70)]
probabilities = [distribution.pdf(value) for value in values]

# plot the histogram and pdf
plt.figure()
plt.hist(sample, bins=10, density=True)
plt.plot(values, probabilities) # line plot
plt.show()
```

#### Distribution 1

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from sklearn.neighbors import KernelDensity

# plot the histogram
plt.figure()
plt.hist(sample1, bins=50)
plt.show()

# fit density
#  - bandwidth = window size / smoothing parameter
#  - kernel = basis function
model = KernelDensity(bandwidth=2, kernel='exponential')
model.fit(sample1)

# sample probabilities for a range of outcomes
values = np.asarray([value for value in range(0, 40)])
values = values.reshape((len(values), 1))
probabilities = model.score_samples(values) # calculate the log probability for the values array
probabilities = np.exp(probabilities) # inverts log operation --> values in range [0,1]

# plot the histogram and pdf
plt.figure()
plt.hist(sample1, bins=50, density=True)
plt.plot(values[:], probabilities)
plt.show()
```

#### Distribution 3

This is actually a bimodal distribution (from two different normal distributions).

Specifically, 300 examples with a mean of 20 and a standard deviation of 5 (the smaller peak), and 700 examples with a mean of 40 and a standard deviation of 5 (the larger peak). The means were chosen close together to ensure the distributions overlap in the combined sample.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
from sklearn.neighbors import KernelDensity

# generate a sample
sample = np.load('distributions/distribution-3.npy')

# plot the histogram
plt.figure()
plt.hist(sample, bins=50)
plt.show()

# fit density
#  - bandwidth = window size / smoothing parameter
#  - kernel = basis function
model = KernelDensity(bandwidth=2, kernel='gaussian')
model.fit(sample)

# sample probabilities for a range of outcomes
values = np.asarray([value for value in range(1, 60)])
values = values.reshape((len(values), 1))
probabilities = model.score_samples(values) # calculate the log probability for the values array
probabilities = np.exp(probabilities) # inverts log operation --> values in range [0,1]

# plot the histogram and pdf
plt.figure()
plt.hist(sample, bins=50, density=True)
plt.plot(values[:], probabilities)
plt.show()
```
